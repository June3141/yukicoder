```python
from math import *
from decimal import *

N = int(input())

c = int(log10(N))
a = Decimal(str(N /(10 ** c))).quantize(Decimal('0.1'), rounding=ROUND_HALF_UP)

if a >= 10:
    a = round(a / 10, 1)
    c += 1

print("{0}*10^{1}".format(a, c))
```
